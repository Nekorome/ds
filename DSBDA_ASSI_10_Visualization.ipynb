```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
def details():
    print("\n\nThe features in the dataset are as follows : ")
    print("1. Id (numeric, ordinal)")
    print("2. Sepal Length (cm) (numeric, Nominal)")
    print("3. Sepal Width (cm) (numeric, Nominal)")
    print("4. Petal Length (cm) (numeric, Nominal)")
    print("5. Petal Width (cm) (numeric, Nominal)")
    print("6. Species (categorical, No. of Categories = 3)\n\n")
```

```python
#histogram
print(sns.distplot(a=df["SepalLengthCm"], bins=[4,5,6,7,8], color="red"))
```

```python
def histogram():
    fig, axes = plt.subplots(2,2, figsize=(10,10), squeeze=False)
    fig.suptitle("Histograms", fontsize=25)
    sns.distplot(a=df["SepalLengthCm"], bins=[0,1,2,3,4,5,6,7,8], color="red", ax=axes[0,0])
    sns.distplot(a=df["SepalWidthCm"], bins=[0,1,2,3,4,5,6,7,8], color="blue", ax=axes[1,0])
    sns.distplot(a=df["PetalLengthCm"], bins=[0,1,2,3,4,5,6,7,8], color="green", ax=axes[0,1])
    sns.distplot(a=df["PetalWidthCm"], bins=[0,1,2,3,4,5,6,7,8], color="black", ax=axes[1,1])
    
    print("\n\nSepal Lengths:")
    print("Sepal Lengths span from around 4cm to 8cm.")
    print("Maximum flowers have a sepal length ranging from 5cm to 6cm.")
    print("\n\nSepal Widths:")
    print("Sepal Widths span from around 2cm to 5cm.")
    print("Maximum flowers have a sepal width ranging from 3cm to 4cm.")
    print("\n\nPetal Lengths:")
    print("Petal Lengths span from around 1cm to 7cm.")
    print("Maximum flowers have a petal length ranging from 1cm to 2cm.")
    print("No flower has a petal length in the range 2cm to 3cm.")
    print("\n\nPetal Widths:")
    print("Petal Widths span from around 0cm to 3cm.")
    print("Maximum flowers have a sepal length ranging from 1cm to 2cm.")
    
    print("\n\nSepal length and Sepal  Width seem to be following a normal distribution.")
    print("Petal length and Petal Width are not following a normal distribution.\n\n")
```

```python
histogram()
```

```python
def boxplots():
    fig, axes = plt.subplots(2,2, figsize=(10,10), squeeze=False)
    fig.suptitle("Box-plots", fontsize=25)
    sns.boxplot(data=df, x="Species", y="SepalLengthCm", ax=axes[0,0])
    sns.boxplot(data=df, x="Species", y="SepalWidthCm", ax=axes[0,1])
    sns.boxplot(data=df, x="Species", y="PetalLengthCm", ax=axes[1,0])
    sns.boxplot(data=df, x="Species", y="PetalWidthCm", ax=axes[1,1])
    
    print("\n\nSepal Length:")
    print("Iris-setosa has the least sepal length amongst the 3 species, followed by Iris-versicolor and finally, Iris-Virginica.")
    print("There is one outlier for Iris-virginica having sepal length lower than its interquartile range of 5.7cm to 7.9cm")
    print("\n\nSepal Width:")
    print("Iris-versicolor has the least Sepal Width amongst the 3 species, followed by Iris-virginica and finally, Iris-Setosa.")
    print("There are two outliers for Iris-virginica having sepal width outside its interquartile range of 2.5cm to 3.6cm")
    print("\n\nPetal Length:")
    print("Iris-setosa has the least Petal Length amongst the 3 species, followed by Iris-versicolor and finally, Iris-Virginica.")
    print("There are three outliers for Iris-setosa having Petal Length outside its interquartile range of 1.2cm to 1.7cm")
    print("There is one outlier for Iris-versicolor having Petal length lower than its interquartile range of 3.2cm to 5.1cm")
    print("\n\nPetal Width:")
    print("Iris-setosa has the least Petal Width amongst the 3 species, followed by Iris-versicolor and finally, Iris-Virginica.")
    print("There are two outliers for Iris-setosa having Petal width outside its interquartile range of 0.2cm to 0.3cm\n\n")
```

```python
boxplots()
```

```python
def main():
    while(True):
        print("1. Details of Dataset")
        print("2. Histograms")
        print("3. Boxplots")
        print("0. Exit")
        ch=int(input("Enter your choice : "))
        if(ch==1):
            details()
        if(ch==2):
            histogram()
        if(ch==3):
            boxplots()
        if(ch==0):
            break
```

```python
main()
```
