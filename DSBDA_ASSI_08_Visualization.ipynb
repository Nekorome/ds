```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
sns.countplot(data = df, x = "Sex")
#count of males and females
```

```python
sns.countplot(data = df, x = "Survived")
#Count of survived and dead
```

```python
sns.countplot(data = df, x = "Survived", hue = "Sex")
#Count of Survived and dead, grouped by sex
```

```python
y=[]
for i in range(891):
    y.append(i)
sns.scatterplot(data=df, x="Age", y=y, hue = "Survived")
# 891 people on y axis and age on x axis, blue is dead and orande is survived
# Finding relation between age and survival
```

```python
y=[]
for i in range(891):
    y.append(i)
sns.scatterplot(data=df, x="Fare", y=y, hue = "Survived")
# 891 people on y axis and age on x axis, blue is dead and orande is survived
# Finding relation between Fare and survival
#Those who paid more survived more
```

```python
sns.distplot(a=df['Age'], bins=4)
#Finding the age distribution of the people
```
